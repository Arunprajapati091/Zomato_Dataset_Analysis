```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
final_df.groupby(['Cuisines','Aggregate rating']).size().reset_index()
```

```python
final_df[final_df['Aggregate rating']<5 ].groupby('Cuisines',sort = True).size().reset_index(name= 'Count').sort_values(by = 'Count',ascending = True)
```

```python
top_cuisines = final_df['Cuisines'].value_counts().head(10)
top_cuisines
```

```python
top_10_cuisines = (
    final_df[final_df['Aggregate rating'] < 5]  # Filter rows with 'Aggregate rating' < 5
    .groupby('Cuisines')  # Group by 'Cuisines'
    .size()  # Count the number of rows in each group
    .reset_index(name='Count')  # Convert to a DataFrame with a 'Count' column
    .sort_values(by='Count', ascending=False)  # Sort by 'Count' in descending order
    .head(10)  # Select the top 10 cuisines
)

print(top_10_cuisines)
```
